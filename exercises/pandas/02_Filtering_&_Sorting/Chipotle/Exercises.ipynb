### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
source_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
dataset = pd.read_csv(source_url, sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = dataset
```

```python
chipo.head(5)
```

### Step _Preprocessing_. Removing duplicate item_name-s
Some entries in the item_name column differ only by occasional hyphens, such as "Chips and Roasted Chili-Corn Salsa" versus "Chips and Roasted Chili Corn Salsa". Iâ€™m standardizing the item_name column by removing hyphens and converting all text to lowercase, in order to merge/remove duplicates. In addition, the item_price parameter is of type str and has to be casted to float64.

```python
chipo['item_name'] = chipo['item_name'].str.replace(r'-', '', regex=True)
chipo['item_name'] = chipo['item_name'].str.lower()

try:
    chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
except TypeError:
    print("Already changed!")
```

### Step 4. How many products cost more than $10.00?

```python
chipo.loc[chipo['item_price'] > 10, 'item_name'].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
just_name_and_price = chipo.groupby('item_name', as_index=False)['item_price'].max()
just_name_and_price
```

### Step 6. Sort by the name of the item

```python
# Was it already sorted?
just_name_and_price.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
price_and_quantity = chipo.groupby('item_name').agg(
    quantity=('quantity', 'sum'),
    price=('item_price', 'max')
).reset_index()

print(price_and_quantity['quantity'].iloc[price_and_quantity['price'].idxmax()])
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
vsb_q = price_and_quantity.loc[
    price_and_quantity['item_name'] == "veggie salad bowl",
    'quantity'
].iloc[0]

print(vsb_q)
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
len(chipo[(chipo['quantity'] >= 2) & (chipo['item_name'] == 'canned soda')])
```
