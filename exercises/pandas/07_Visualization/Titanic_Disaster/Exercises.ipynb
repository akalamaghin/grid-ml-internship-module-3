# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

```python
source_url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv"
dataset = pd.read_csv(source_url)
```

### Step 3. Assign it to a variable titanic 

```python
titanic = dataset
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
labels = ["Male", "Female"]
counts = titanic['Sex'].value_counts()
parts = [
    counts[sex.lower()] / counts.sum() for sex in labels
]

colors = ["blue", "red"]

plt.pie(
    parts,
    colors=colors,
    startangle=90,
    autopct='%1.1f%%',
    textprops={'color':"white", 'fontsize':12}
)

plt.legend(labels, title="Sex", loc="upper right")
plt.title("Gender Proportion")

plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
color_map = {"male": "blue", "female": "red"}
sex_color = lambda x: color_map[x]

plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.4, color=titanic['Sex'].apply(sex_color))

plt.xlabel('Age')
plt.ylabel('Fare')

for sex, color in color_map.items():
    plt.scatter([], [], color=color, label=sex)

plt.legend(title="Sex")
plt.show()
```

### Step 7. How many people survived?

```python
labels = ['No', 'Yes']
counts = titanic['Survived'].value_counts()
parts = [
    counts[i] / counts.sum() for i in range(2)
]

colors = ["red", "blue"]

plt.pie(
    parts,
    colors=colors,
    startangle=90,
    autopct='%1.1f%%',
    textprops={'color':"white", 'fontsize':12}
)

plt.legend(labels, title="Survived", loc="upper right")
plt.title("Survivors Proportion")

plt.show()
```

### Step 8. Create a histogram with the Fare payed

```python
titanic['Fare'].plot(kind='hist', bins=20, color="blue")

plt.xlabel('Fare')

plt.show()
```
