### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

```python
source_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
dataset = pd.read_csv(source_url, sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo = dataset
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top5 = chipo['item_name'].value_counts().iloc[:5]

top5.plot(kind='bar')
plt.xlabel('Item Name')
plt.ylabel('Count')
plt.title('Top 5 Most Frequent Items')
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

```python
orders = chipo.groupby('order_id')[['quantity', 'item_price']].sum()
```

```python
plt.scatter(x = orders.item_price, y = orders.quantity, s = 50)

plt.xlabel('Order Price')
plt.ylabel('Items ordered')
plt.title('Number of items ordered per order price');
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
import matplotlib.pyplot as plt

orders['item_price'].plot(kind='hist', bins=50)
plt.xlabel('Order Price')
plt.ylabel('Frequency')
plt.title('Distribution of Order Prices')
plt.show()
```
