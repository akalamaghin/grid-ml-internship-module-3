### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
source_url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
dataset = pd.read_csv(source_url, sep='|')
```

### Step 3. Assign it to a variable called users.

```python
users = dataset
users.head(10)
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
gender_counts_per_occupation = users.groupby(['occupation', 'gender']).size()
total_per_occupation = users.groupby('occupation').size()
male_counts = gender_counts_per_occupation.xs('M', level='gender')
male_ratios = male_counts / total_per_occupation
male_ratios = male_ratios.sort_values(ascending=False)
male_ratios.name = 'ratio'

male_ratios
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
female_counts = gender_counts_per_occupation.xs('F', level='gender')
female_ratios = female_counts / total_per_occupation
female_ratios = female_ratios.sort_values(ascending=False)
female_ratios.name = 'ratio'

df = pd.concat([male_ratios, female_ratios], keys=['M', 'F'], names=['gender', 'occupation']).reset_index()
df = df.set_index(['occupation', 'gender'])
df = df.sort_values(['occupation', 'gender'])

df['ratio'] = df['ratio'].fillna(0)
df['%'] = df['ratio'].apply(lambda x: f"{x * 100:.2f}")

df[['%']]
```
